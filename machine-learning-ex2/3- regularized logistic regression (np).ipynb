# feature mapping

polynomial expansion

```
for i in 0..i
  for p in 0..i:
    output x^(i-p) * y^p
```
<img style="float: left;" src="../img/mapped_feature.png">

````python
x1 = np.array(df.test1)
x2 = np.array(df.test2)
````

````python
data = lr.feature_mapping(x1, x2, power=6)
print(data.shape)
data.head()
````

# regularized cost
<img style="float: left;" src="../img/reg_cost.png">

````python
theta = np.zeros(data.shape[1])
X = lr.feature_mapping(x1, x2, power=6, as_ndarray=True)
print(X.shape)

y = general.get_y(df)
print(y.shape)
````

````python
lr.regularized_cost(theta, X, y, l=1)
````

this is the same as the not regularized cost because we init theta as zeros...

# regularized gradient
<img style="float: left;" src="../img/reg_gradient.png">

````python
lr.regularized_gradient(theta, X, y)
````

# fit the parameters

````python
import scipy.optimize as opt
````

````python
print('init cost = {}'.format(lr.regularized_cost(theta, X, y)))

res = opt.minimize(fun=lr.regularized_cost, x0=theta, args=(X, y), method='Newton-CG', jac=lr.regularized_gradient)
res
````

# predict

````python
final_theta = res.x
y_pred = lr.predict(X, final_theta)
accuracy = np.mean(y_pred == y)

print(accuracy)
````

# regularization and feature mapping effect

````python
power_max, l_max = 20 ,20

z = np.array([[lr.accuracy(x1, x2, y, power, l) for power in np.arange(1,power_max+1)] 
                  for l in np.arange(1,l_max+1)])

power, l = np.meshgrid(np.arange(1, power_max+1), np.arange(1, l_max+1))
````

````python
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm

fig = plt.figure(figsize=(16,10))
ax = fig.gca(projection='3d')

surf = ax.plot_surface(power, l, z, rstride=1, cstride=1, cmap=cm.coolwarm, linewidth=0, antialiased=False)
ax.set_xlabel('power')
ax.set_ylabel('L constant')

fig.colorbar(surf, shrink=0.5, aspect=5)
````
