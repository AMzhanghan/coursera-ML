```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

# prepare data

```python
df = pd.read_csv('ex2data1.txt', names=['exam1', 'exam2', 'admitted'])
print(df.shape)
df.head()
```

```python
df.describe()
```

# pandas plot

```python
df.plot.scatter(x='exam1', y='exam2', c='admitted', figsize=(10,10),colorbar=False)
```

# matplotlib

```python
fig, ax = plt.subplots(figsize=(10, 10))

ax.scatter('exam1', 'exam2', data=df, c='admitted', s=100)
```

# seaborn

```python
sns.set(context="notebook", style="white")
sns.set_palette('bright')

sns.lmplot('exam1', 'exam2', hue='admitted', data=df, 
           size=9, 
           fit_reg=False, 
           scatter_kws={"s": 100}
          )
```
